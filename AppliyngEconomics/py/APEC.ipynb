```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

gdps =[3.38e+10,3.36e+10,7.93e+10,2.18e+11,1.03e+11,7.21e+11,3.38e+11]

growths = [2.116631,1.878938,2.312074,3.966353,3.072762,2.563569,4.713699 ]

exps= [3.58975,3.841845,4.672633,4.895133,3.128302,3.684973,3.111009]

quality = [431.9202,476.5847,486.8729,501.5589,428.9178,474.4484,444.7606]


countries = ["Bulgaria","Croatia","Hungary","Poland","Romania","Russian Federation","Turkey"]
titles = ["Country","GDP","Growth","Education Expenditure","Quality"]

plt.figure(figsize=(11,8))
for i in range(7):
  plt.scatter(exps[i], quality[i], label = countries[i],s=200)


x = []
y = []

for i in range(7):
  if countries[i] == "Poland" or countries[i] == "Russian Federation" or countries[i] == "Turkey":
    x.append(exps[i])
    y.append(quality[i])

plt.plot(x,y)

plt.ylabel('PISA Score (Average of All Subjects)',fontsize=13)
plt.xlabel("Education Expenditure (% Share of GDP)",fontsize=13)
plt.title("Education Efficiency (Expenditure - Quality)",fontsize=18)
plt.legend(loc=4, prop={'size': 12})
plt.grid()
plt.show()
```
